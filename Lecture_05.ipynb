## Simplicial homology

**Definition:** Let $X$ be a simplicial complex and let $R$ be a ring. The *module of simplicial $k$-chains* of $X$ with coefficients in $R$ is defined as
$$
C_k(X;R) := \bigoplus_{\sigma \in X, \operatorname{dim}(\sigma)=k} R
$$
The elements $\sum\limits_{\sigma\in X, \operatorname{dim}(\sigma)=k}r_\sigma$ are called $k$-chains.

We are interested in three particular cases: $R =\mathbb{Z}$, $R=\mathbb{F}$ for a field $\mathbb{F}$, and, in particular, $R=\mathbb{Z}/2$.

## Simplicial homology

For $R=\mathbb{Z}$ and $R=\mathbb{Z}/2$ chains have a nice geometric interpretation:

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

apoints = np.array([[0,0],[1,1],[2,0],[2,-1.5],[1,-2.5],[0,-1.5],[0,0]])
bpoints = np.array([[0,0],[0,1.5],[1,1],[2,1.5],[2,0],[2.75,-0.75],[2,-1.5],[2,-3],[1,-2.5],[0,-3],[0,-1.5],[-0.75,-0.75],[0,0]])
cpoints = np.array([[0,1.5],[2,1.5],[2.75,-0.75],[2,-3],[0,-3],[-0.75,-0.75],[0,1.5]])
xapoints,yapoints  = apoints.T
xbpoints,ybpoints  = bpoints.T
xcpoints,ycpoints  = cpoints.T
ax = plt.gca()
ax.set_xlim([-2, 4])
ax.set_ylim([-5, 3])

t1= plt.Polygon(cpoints,color='purple')
plt.gca().add_patch(t1)
t2= plt.Polygon(apoints,color='white')
plt.gca().add_patch(t2)
#t3=plt.Polygon([[2,0],[2.75,-0.75],[2,-1.5]],color='purple')
#plt.gca().add_patch(t3)
plt.plot(xapoints,yapoints,linewidth='4',color='green')
plt.plot(xbpoints,ybpoints,color='black')
plt.plot(xcpoints,ycpoints,linewidth='4',color='green')
#plt.plot([-0.75,0],[-0.75,1.5],color='green',linewidth='4')
#plt.plot([0,2],[-3,-3],color='green',linewidth='4')
plt.plot(xapoints,yapoints,'o',color='blue')
plt.plot(xbpoints,ybpoints,'o',color='blue')
plt.plot(xcpoints,ycpoints,'o',color='blue')

plt.savefig('05_05')
```
